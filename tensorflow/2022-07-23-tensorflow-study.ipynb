## 컴파일 과정에서 나오는 loss optimizer metrics 의 의미
> ### loss 손실 함수 ( 오차 측정 방법 )
> ##### - 예측값과 실제 값을 비교하며 훈련 중에 오차가 얼마나 있는지를 측정한다. 이 함수를 최소화를 해야함<br>
> #### 종류
> ##### 다중 분류 
> 이거는 sparse_categorical_crossentropy 하고는 다르게 원 핫 인코딩 함 [1,0,0] [0,1,0] [0,0,1] sparse_catogrical_crossentropy 는 레이블을 정수 형태로 사용가능
ex) Lable : 1,2,3
> ```
> model.compile( optimiszers='Adam',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'] )
> ```
> ##### 이진 분류
> ```
> model.compile( optimiszers='Adam',
                 loss='binary_crossentropy',
                 metrics=['accuracy'] )
> ```
> ##### 회귀
> ```
> model.compile( optimiszers='Adam',
                 loss='mse', # 또는 mae 사용
                 metrics=['accuracy'] )
> ```

> ### optimizer 최적화 알고리즘 ( 훈련하는 동안 최소화 시켜야 함 )
> ##### - 종류가 많음 SDG,Adagrad , RMSProp, Adam 등 여기서 나오는 개념이 `경사하강법`이라고 있음
> 자세한건 생략 근데 일반적으로 Adam 이 성능이 좋아서 Adam을 많이 사용함

> ### metrics ( 지표 ) 오차를 평가하기 위한 방법 ? 
> #### 분류 : accuracy 사용
> #### 회귀 : MAE,MES 등 사용
> ##### MAE ( Mean Absolute Error ) : 실제 값과 예측 값 차이를 `절대값`으로 변환하여 평균한 것
> ##### MES ( Mean Squared Error ) : 실제 값과 예측 값 차이를 `제곱`하여 평균한 것

### 베이스 모델 돌리기

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Flatten

model = Sequential()
model.add( Flatten( input_shape=( 28,28 ) ) )
model.add( Dense( 64 , activation='relu' ) )
model.add( Dense( 10 , activation='softmax' ) )

model.compile( optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'] )
````

### 학습 하는 방법 model.fit()을 하게 되면 학습이 가능함
> #### model.fit(x_train, y_train)
> ##### x_train : Feature(특징?)<br>
> ##### y_train : label(레이블)
> `mnist` : 숫자 손글씨로 작성된 데이터셋임 리턴값이 2개 (아마도) train 하고 test 셋이 있는 걸로 암

###  ⭐참고⭐
#### X or x 가 들어가면 Feature(피처) (기능이라고 보면됨)
#### Y or y 가 들어가면 lable 혹은 target이라고 이해하면 편함
#### train은 훈련용 데이터 셋     test는 검증용 데이터 셋 

#### 데이터셋 불러오기

````python
mnist = tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test) = mnist.load_data()
# x_train => (60000 , 28 , 28 )형태
# y_train => (60000 ,)
# x_test  => (10000 , 28, 28,)
# y_test  => (10000 ,)
````

#### 데이터 셋 보기 ( 손 글씨 출력 )

````python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])         # 60000 만개의 데이터 셋 중에 0번째
print("lable:",y_train[0])     # 레이블
````

### 학습 하기전 용어 정리
```
Epoch 1/3
1875/1875 [==============================] - 4s 2ms/step - loss: 2.1948 - accuracy: 0.8064
Epoch 2/3
1875/1875 [==============================] - 4s 2ms/step - loss: 0.4187 - accuracy: 0.8915
Epoch 3/3
1875/1875 [==============================] - 3s 2ms/step - loss: 0.3233 - accuracy: 0.9170
```
#### epoch(에폭시) : 학습을 몇 번 할건지
> 학습을 할 때 데이터 전부를 한꺼번에 학습하는게 아님 `데이터 셋` 을 `배치 사이즈`로 나눠서 함 그래서 저기 ePOCH 밑에 쓰여져 있는 1875는 `배치사이즈`를 몇 번 돌릴 지를 의미하는데 배치사이즈 default 값이 32 이기 때문에(데이터 셋 60000) 32 * 1875 = 60000 | 32 정도의 배치 사이즈를 1875번 돌리고 그것을 총 epoch 3 3번 돌린다는 의미가 됨
#### batch_size : 한번의 배치마다 전체 데이터에서 일부를 불러오는 사이즈
#### Iteration : 한 epoch당 필요한 배치 사이즈 개수 == 파라미터 업데이트 개수
#### loss : 손실도 ( 작음 좋음 )
#### accuracy : 정확도 ( 크면 좋음 )

````python
history = model.fit(x_train,y_train, epochs=5, batch_size=64)
# 참고로 verbose=0으로 하면 학습 진행도가 안뜸 == 데이터셋이 크면 이거 로딩을 하는데 시간이 걸릴 수도 있기때문
````

````python
# 학습 정확도 확인
plt.plot(history.history['accuracy'])
plt.xlabel('epochs')
plt.ylabel('accuracy')
````

# 두번 째 방법

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

mnist = tf.keras.datasets.mnist
(X,y),(x_test,y_test) = mnist.load_data()

from sklearn.model_selection import train_test_split
x_train , x_val , y_train , y_val = train_test_split(X,y,test_size=0.2)

model = Sequential([
                    Flatten(input_shape=(28,28)),
                    Dense(64,activation='relu'),
                    Dense(10,activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit( x_train , y_train , validation_data=( x_val , y_val ) , epochs=5 , batch_size=64)
````

````python
# 베이스 모델
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Flatten

mnist = tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test) = mnist.load_data()

model = Sequential()
model.add( Flatten( input_shape=( 28,28 ) ) )
model.add( Dense( 64 , activation='relu' ) )
model.add( Dense( 10 , activation='softmax' ) )

model.compile( optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'] )

history = model.fit( x_train , y_train ,validation_split = 0.2 , epochs = 5 , batch_size=64 )
````
